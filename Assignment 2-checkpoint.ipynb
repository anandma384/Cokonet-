## Creating & Manipulating Dataframes

```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings("ignore")

names = ["John", "Mary", "Peter"]
age = [27, 34, 47]
sex = ["male", "female", "female"]

data = { "Name": names, "Age": age, "Sex": sex}

New_dataframe = pd.DataFrame(data)
print(New_dataframe)
```

### Using the .loc attribute, access the age of John from the DataFrame.

```python
John_Age = New_dataframe.loc[0,'Age']    # 0 for John, 1 for Mary & 2 for Peter
print("Age of John is ", John_Age)
```

### Using the .loc attribute change the gender of Peter from "female" to "male."

```python
New_dataframe.loc[New_dataframe['Name'] == 'Peter', 'Sex'] = 'male'
print("DataFrame after gender change:")
print(New_dataframe)
```

### Write code to extract a subset of the DataFrame containing only male individuals.

```python
Male_dataframe = New_dataframe[New_dataframe['Sex'] == 'male']
Male_dataframe
# Or else assign New_dataframe['Sex'] == 'male' with any variable name and assign
```

### Write another code to retrieve the name and age of the oldest male from the modified DataFrame.

```python
max_age_per_sex = Male_dataframe.groupby('Sex')['Age'].max()

oldest_male = max_age_per_sex['male']
print("The oldest male age is", oldest_male)


```
